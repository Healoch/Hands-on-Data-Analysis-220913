```python
import pandas as pd
import numpy as np
```

```python
text = pd.read_csv('train_chinese.csv',encoding='GBK')
text.head(3)
```

```python
#自己构建一个DataFrame
test0 = pd.DataFrame({'x':pd.Series([4,2,1], index=['c','b','a']),
                      'z':pd.Series([8,6,5], index=['c','b','a']),
                      'y':pd.Series([9,7,3], index=['c','b','a'])})
test0
```

```python
test0.sort_values('x') #根据x列升序排序
```

```python
#让行索引升序排序
test0.sort_index()
```

```python
#让列索引升序排序
test0.sort_index(axis=1)
```

```python
#让列索引降序排序
test0.sort_index(axis=1,ascending=False)
```

```python
#任选两列数据同时降序排序
test0.sort_values(by=['x','z'],ascending=False)
```
